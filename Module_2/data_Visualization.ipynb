#Visualization

```python
import matplotlib.pyplot as plt
a = ['tea', 'cofee', 'coke', 'fanta', 'pepsi']
b = [23, 45, 12, 23, 90]

plt.plot(a, b, color = 'red')
plt.title('drinks people take')
plt.xlabel('persons')
plt.ylabel('numbers')
plt.show()
```

```python
#scatter plot
plt.scatter(a, b, color = 'magenta')
plt.title('drinks people take')
plt.xlabel('persons')
plt.ylabel('numbers')
plt.show()
```

```python
#barchart
plt.figure(figsize=(10,5))
plt.bar(a, b, color = 'brown')
plt.title('drinks people take')
plt.xlabel('persons')
plt.ylabel('numbers')
plt.show()
```

```python
#histogram

plt.hist(b, color = 'orange')
plt.title('drinks people take')
plt.xlabel('persons')
plt.ylabel('numbers')
plt.show()
```

VISUALIZATING WEB3 TRANSFER DATA

```python
df = pd.read_csv(r'c:\Users\Muhammad\Blockchain_Assignments\web3_token_transfers.csv', parse_dates = ['timestamp'])
df['date'] = df['timestamp'].dt.date

#set seaborn theme
sns.set(style ='whitegrid')
df.head(5)
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['amount'], bins=50, kde=True, color='skyblue')
plt.title('Distribution of Token Transfer Amounts')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.xlim(0, 3000)
plt.show()
```

```python
#daily total transfer

daily_volume = df.groupby('date')['amount'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data = daily_volume, x='date', y='amount', marker= 'o')
plt.title('Daily total transfer volume') 
plt.xlabel('date')
plt.ylabel('Total Amount Transferred')
plt.tight_layout()
plt.show()
```
